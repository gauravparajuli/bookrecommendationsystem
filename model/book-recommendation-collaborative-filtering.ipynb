```python
ratings_with_names = ratings.merge(books, on='ISBN')
num_ratings_df = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df = num_ratings_df.rename(columns={'Book-Rating': 'num_ratings'})
num_ratings_df
```

```python
avg_ratings_df = ratings_with_names.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df = avg_ratings_df.rename(columns={'Book-Rating': 'avg_ratings'})
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

```python
popular_df
```

```python
import pickle
with open('top50books.pkl', 'wb') as f:
    pickle.dump(popular_df, f)
```

# collaborative filtering

```python
ratings_with_names.head(2)
```

```python
x = ratings_with_names.groupby('User-ID').count()['Book-Rating'].reset_index()
```

```python
users_reviews_gt_200 = x[x['Book-Rating'] > 200]['User-ID'].to_list()
```

```python
users_reviews_gt_200
```

```python
filtered_ratings = ratings_with_names[ratings_with_names['User-ID'].isin(users_reviews_gt_200)]
```

```python
filtered_ratings
```

```python
# for books
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] > 50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0].shape
```

```python
similarity_score.shape
```

```python
def recommend(bookname):
    # fetch book index
    index = np.where(pt.index == bookname)[0][0]
    similiar_items = sorted(list(enumerate(similarity_score[index])), key= lambda x: x[1], reverse=True)[1:8]
    
    data = []
    for i in similiar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
    
```
